```python
from chatterbot import ChatBot
from chatterbot.trainers import ChatterBotCorpusTrainer

# Create a new chatbot
college_bot = ChatBot('CollegeBot')

# Create a new Trainer and set it to train the chatbot on the English language corpus data
trainer = ChatterBotCorpusTrainer(college_bot)
trainer.train('chatterbot.corpus.english')

# Custom data for college-related information
college_data = {
    'name': 'smbst college sangamnnear',
    'location': 'sangamnear, india',
    'courses': ['Computer Science', 'Electrical Engineering', 'Business Administration'],
    'clubs': ['Coding Club', 'Music Club', 'Sports Club'],
    'events': ['Tech Fest', 'Cultural Night', 'Sports Day']
}

# Function to provide college-related information
def get_college_info(query):
    query = query.lower()
    if 'name' in query:
        return college_data['name']
    elif 'location' in query:
        return college_data['location']
    elif 'courses' in query:
        return ', '.join(college_data['courses'])
    elif 'clubs' in query:
        return ', '.join(college_data['clubs'])
    elif 'events' in query:
        return ', '.join(college_data['events'])
    else:
        return "I'm sorry, I don't have information on that."

# Function to interact with the chatbot
def chat_with_bot():
    print("College Bot: Hello! I'm your College Bot. How can I help you today?")
    while True:
        user_input = input("You: ")
        if user_input.lower() == 'exit':
            print("College Bot: Goodbye!")
            break
        elif user_input.lower() == 'help':
            print("College Bot: You can ask me about the college name, location, courses, clubs, or events.")
        else:
            response = college_bot.get_response(user_input)
            print(f"College Bot: {response}")

        # Check if the user is asking for specific college information
        college_info = get_college_info(user_input)
        if college_info:
            print(f"College Bot: {college_info}")

if __name__ == "__main__":
    chat_with_bot()
```

```python
# Constants for player representation
PLAYER_X = 'X'
PLAYER_O = 'O'
EMPTY = ' '

def evaluate(board):
    # Evaluate the current state of the board
    if check_winner(board, PLAYER_X):
        return 10
    elif check_winner(board, PLAYER_O):
        return -10
    else:
        return 0

def is_move_left(board):
    # Check if there are empty cells left on the board
    return any(cell == EMPTY for cell in board)

def check_winner(board, player):
    # Check if the player has won
    win_conditions = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # Rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # Columns
        [0, 4, 8], [2, 4, 6]              # Diagonals
    ]
    for condition in win_conditions:
        if all(board[i] == player for i in condition):
            return True
    return False

def minimax(board, depth, is_maximizer):
    score = evaluate(board)

    if score == 10:
        return score

    if score == -10:
        return score

    if not is_move_left(board):
        return 0

    if is_maximizer:
        best = -float('inf')
        for i in range(9):
            if board[i] == EMPTY:
                board[i] = PLAYER_X
                best = max(best, minimax(board, depth + 1, not is_maximizer))
                board[i] = EMPTY
        return best
    else:
        best = float('inf')
        for i in range(9):
            if board[i] == EMPTY:
                board[i] = PLAYER_O
                best = min(best, minimax(board, depth + 1, not is_maximizer))
                board[i] = EMPTY
        return best

def find_best_move(board):
    best_val = -float('inf')
    best_move = -1

    for i in range(9):
        if board[i] == EMPTY:
            board[i] = PLAYER_X
            move_val = minimax(board, 0, False)
            board[i] = EMPTY

            if move_val > best_val:
                best_move = i
                best_val = move_val

    return best_move

def print_board(board):
    for i in range(0, 9, 3):
        print(f"{board[i]} | {board[i + 1]} | {board[i + 2]}")
    print()

if __name__ == "__main__":
    board = [EMPTY] * 9

    print("Initial Board:")
    print_board(board)

    while is_move_left(board):
        # Player's move
        player_move = int(input("Enter your move (0-8): "))
        if board[player_move] != EMPTY:
            print("Invalid move. Cell already taken. Try again.")
            continue
        board[player_move] = PLAYER_O

        print("Player's Move:")
        print_board(board)

        if check_winner(board, PLAYER_O):
            print("You win!")
            break

        if not is_move_left(board):
            print("It's a tie!")
            break

        # Computer's move
        computer_move = find_best_move(board)
        board[computer_move] = PLAYER_X

        print("Computer's Move:")
        print_board(board)

        if check_winner(board, PLAYER_X):
            print("Computer wins!")
            break

        if not is_move_left(board):
            print("It's a tie!")
            break
```
